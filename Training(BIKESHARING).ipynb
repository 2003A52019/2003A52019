```python
df=df.rename({'dteday':'dateday'},axis=1)
df.columns
```

```python
#replacing numeric values to categorical(season,weathersit,weekday)
df1 = {"season" : {1:"winter", 2:"spring", 3:"summer", 4:"fall"}, 
             "weathersit" : {1: "Clear/Partly Cloudy", 2: "Mist and Cloudy", 3: "Light Snow/Rain"},
             "weekday" : {0: "Sunday", 1:"Monday", 2:"Tuesday", 3:"Wednesday", 4:"Thursday", 5:"Friday", 6:"Saturday"}}
df=df.replace(df1)
print(df)
```

```python
#bar chart using matplotlib
plt.figure(figsize=(6,6))
plt.title('season vs count')
plt.xlabel("season")
plt.ylabel("count")
plt.bar(df.season,df.cnt)
plt.show()
```

```python
#bar graph using sea born
plt.figure(figsize=(8,8))
#Creating bar plot using seaborn
sns.barplot(x = 'season',y = 'cnt',data = df,palette = "Oranges")
plt.title('Bike sharing based on Weathersit')
plt.xlabel('Weathersit ')
plt.ylabel('count') 
plt.show()
```

```python
sns.barplot(x = 'yr',y = 'mnth',hue = 'holiday',data = df, palette = "Greens")
plt.title('Comparative Ananlysis')
plt.xlabel('yr')
plt.ylabel('mnth')
plt.show()
```
